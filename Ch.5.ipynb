```python
import math
import numpy as np
import pandas as pd
import random
import scipy.stats as stats 
import matplotlib.pyplot as plt
```

```python
# equation 5.1 relates to example in ch3
# Randomly select two people w/ combined wealth of $36m
# The most likely combination is not $18m and $18m, but rather $35.999m and $1000
```

```python
# Inventing a distribution on cusp of Gaussian and subexponential
# i.e. thin to fat tails
```

```python
x = np.linspace(-5,5,1000)

# same as 1/2*k*sech^2(kx)
k = 1
y = .5*k/(np.cosh(x*k))

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the invented function
plt.plot(x,y, 'r')

# plot normal distrib
plt.plot(x, stats.norm.pdf(x, 0, 1), 'b')

plt.show()
```
