```python
# pg 282, equation 15.9
def quantileWithMax(p, alpha, sigma, H, L):
    gamma = (alpha * sigma * (1 - p) ** (-1 / alpha)) / H
    q = math.exp(gamma) * (
        L * math.exp(alpha * sigma / H)
        + H * math.exp(gamma)
        - H * math.exp(alpha * sigma / H)
    )
    return q
```
